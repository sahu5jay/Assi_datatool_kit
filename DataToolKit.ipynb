```python
# Que-26 How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    'Age': [25, 30, 35, 40, 28],
    'Salary': [50000, 60000, 70000, 80000, 58000],
    'Experience': [2, 5, 8, 10, 3]
}

df = pd.DataFrame(data)

corr_data = df.corr()

plt.figure(figsize=(8,6))
sns.heatmap(data = corr_data, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()

```
